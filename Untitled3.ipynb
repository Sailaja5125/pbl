```python
%%writefile app.py
import streamlit as st
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
import numpy as np
import matplotlib.pyplot as plt

# Function to fetch headlines from a URL
def fetch_headlines(url):
    response = requests.get(url)
    headlines = []
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Scraping headlines from the top-newslist
        top_news_div = soup.find('div', class_='top-newslist')
        if top_news_div:
            news_cards = top_news_div.find_all('li')
            for card in news_cards:
                headline = card.find('a', title=True)
                if headline:
                    headlines.append(headline['title'])

        # Scraping headlines from the headlines-list
        headlines_div = soup.find('div', class_='headlines-list')
        if headlines_div:
            news_cards = headlines_div.find_all('li')
            for card in news_cards:
                headline = card.find('a', title=True)
                if headline:
                    headlines.append(headline['title'])
    return headlines

# Function to analyze sentiment of headlines
def analyze_sentiment(headlines):
    return pipeline('sentiment-analysis')(headlines)

# Function to create a pie chart of sentiment analysis results


def create_pie_charts(data):
    # Separate the scores based on the label
    positive_scores = [entry['score'] for entry in data if entry['label'] == 'POSITIVE']
    negative_scores = [entry['score'] for entry in data if entry['label'] == 'NEGATIVE']

    # Calculate the average scores
    average_positive_score = np.mean(positive_scores)
    average_negative_score = np.mean(negative_scores)

    # Calculate the counts
    positive_count = len(positive_scores)
    negative_count = len(negative_scores)

    # Create the pie chart for counts
    labels = ['Positive', 'Negative']
    counts = [positive_count, negative_count]
    colors = ['blue', 'red']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    ax1.set_title('Distribution of Labels')

    # Create the pie chart for average scores
    average_scores = [average_positive_score, average_negative_score]

    ax2.pie(average_scores, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    ax2.set_title('Average Positivity')

    return fig



# Streamlit app
st.title("News Sentiment Analysis")

# URL input
url = st.text_input("Enter the URL of the news website:", "https://timesofindia.indiatimes.com/home/headlines")

# Fetch headlines button
if st.button("Fetch Headlines"):
    data = []
    headlines = fetch_headlines(url)
    if headlines:
        st.write("**Headlines:**")
        for headline in headlines:
            st.write(headline)
            data.append(headline)

        data2 = []
        for d in analyze_sentiment(data):
              data2.append(d)
        st.write("**Sentiment Analysis:**")
        for d1 in data2:
            st.write(d1)

        # Display pie chart
        fig = create_pie_charts(data2)
        st.pyplot(fig)
        plt.show()  # Display the pie charts

    else:
        st.write("No headlines found for this URL.")
```
